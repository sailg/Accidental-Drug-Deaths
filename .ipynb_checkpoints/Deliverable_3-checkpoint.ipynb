## "Which age group are facing deaths at most?"

```python
plt.figure(figsize=(20,10))
bins= [15,25,35,45,55,65,75,85]
plt.hist(df.Age, bins=bins, edgecolor="k")
plt.xticks(bins,fontsize=20)
plt.yticks(fontsize=20)
plt.xlabel("Age Intervals",fontsize=30)
plt.ylabel("Number of deaths",fontsize=30)
my_colors = 'rgbkymc'
plt.show()
```

People between age 45-55 are facing lot of deaths due to drugs. 25-35 age group comes next with a slight difference in accidental deaths due to drugs

```python
df.head(1)
```

## "Which gender group facing more deaths?"

```python
sns.countplot(x="Sex", data=df)
```

From this graph we can clearly say that Males are facing more accidental deaths due to drugs than females.

### "Which drug is reason for more deaths ?"

```python
list_drugs=list(df.columns[19:33]) # creating a list of drugs
```

```python
drug_dict={}
for drug in list_drugs:
    count=df[drug].notna().sum()
    drug_dict[drug]=count
drug_dict
```

```python
# Plotting Drug deaths Graph
plt.figure(figsize=(20,10))
names = list(drug_dict.keys())
values = list(drug_dict.values())
plt.bar(range(len(drug_dict)),values,tick_label=names) #tick_label does the some work as plt.xticks()
plt.xlabel("Drug type",fontsize=20)
plt.ylabel('Number of deaths',fontsize=20)
plt.yticks(fontsize=20)
plt.xticks(fontsize=20,rotation=90)
plt.show()
```

Heroine and Cocaine are the drugs that caused more accidental deaths
